```python
df_raw.drop_duplicates(subset=["Property_ID"],inplace= True) #drop this propert id because 1lk unique value
```

```python
#df_raw["Ticket_Price_Cr"] = df_raw["Ticket_Price_Cr"].fillna(df_raw["Ticket_Price_Cr"].median())
```

```python
df_raw["Ticket_Price_Cr"] = df_raw["Ticket_Price_Cr"].fillna(0)
```

```python
(df_raw["Ticket_Price_Cr"] == 0).sum()
print(df_raw["Ticket_Price_Cr"].head)
```

```python

df_raw["Ticket_Price_Cr"] = (
    df_raw["Ticket_Price_Cr"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(" Cr", "", regex=False) # regex = Flase means no use this Pattern
)
```

```python
df_raw["Ticket_Price_Cr"] = pd.to_numeric(df_raw["Ticket_Price_Cr"], errors="coerce") #errors="coerce means not able to covert mention NaN
```

```python
df_raw["Ticket_Price_Cr"].isnull().sum()
```

```python
Cat_data = df_raw.select_dtypes(include=object)
```

```python
for col in Cat_data.columns:
    print(df_raw[col].value_counts())
```

```python
for col in Cat_data:
    df_raw[col] = df_raw[col].str.lower().str.strip().str.title() #change same format
```

```python
cat_cols = df_raw.select_dtypes(include='object').columns # space + small spelling mismatch fix

for col in cat_cols:
    df_raw[col] = (
        df_raw[col]
        .str.replace(r"\s+", " ", regex=True)
        .str.title()
    )
```

```python
df_raw.describe()
```

```python
median_price = df_raw["Ticket_Price_Cr"].median()
df_raw.loc[df_raw["Ticket_Price_Cr"] < 0, "Ticket_Price_Cr"] = median_price #fix min ticket price nagative value
```

```python
#Feature Engineering
```

```python
df_raw['Quarter_Number'] = df_raw['Purchase_Quarter'].dt.quarter  #2023 Q1 sales : add quater colums 
```

```python
df_raw['Price_per_Sqft'] = (df_raw['Ticket_Price_Cr'] * 10000000) / df_raw['Unit_Size_Sqft'] #property valuation {Ticket_Price_Cr×1,00,00,000​ % unit sqft}
print(df_raw["Price_per_Sqft"]) #add price_per_sqft colum
```

```python
df_raw['Booking_Flag'] = df_raw['Ticket_Price_Cr'].apply(
    lambda x: 1 if x > 0 else 0 #add booking flag colum booked means 1 if not 0
)
```

```python
df_raw["Booking_Flag"].value_counts()
```

```python
df_raw.drop('NRI_Buyer', axis=1, inplace=True) #NRI_Buyer = Buyer_Type showing this same detailes
```

```python
df_raw["Booking_Status"] = df_raw["Ticket_Price_Cr"].apply(
    lambda x: "Booked" if x > 0 else "Not Booked"
)
```

```python
from sqlalchemy import create_engine

engine = create_engine("postgresql://postgres:root@localhost:5432/mydb")

df_raw.to_sql("luxury_housing", engine, if_exists="replace", index=False)

print("✅ Data inserted into SQL")
print(engine)
```
